```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
#reading csv data

df=pd.read_csv("HR_Employee_Attrition.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#converting Y (Attrition) into numbers
df["Attrition"].replace(to_replace='Yes', value=1, inplace=True)
df["Attrition"].replace(to_replace='No', value=0, inplace=True)
df
```

```python
#convert gender into numbers
df["Gender"].replace(to_replace='Female', value=1, inplace=True)
df["Gender"].replace(to_replace='Male', value=0, inplace=True)
df["Gender"].head()
```

```python
# checking the sum of number of missing  values
df.isnull().sum()
```

```python
#converting all the categorical variables into dummy variables
df_dummies = pd.get_dummies(df)
df_dummies.head()
```

```python
df_dummies.info()
```

```python
# Correlation of "Attrition" with other variables:
# figure size 16 width and 9 height 
plt.figure(figsize=(16,9))

# bar chart displaying  correration  and sort from highest feature that  affect Attrition 

df_dummies.corr()['Attrition'].sort_values(ascending = False).plot(kind='bar')
```

```python
#Average monthly income by education and attrition.
distancefromhome_JobRole_attrition = df[["DistanceFromHome", "Attrition", "JobRole"]]
distancefromhome_JobRole_attrition= distancefromhome_JobRole_attrition.groupby(["JobRole", "Attrition"]).agg('mean')
print(distancefromhome_JobRole_attrition)
distancefromhome_JobRole_attrition.unstack(level=-1)['DistanceFromHome'].plot(ylabel='DistanceFromHome',kind='bar')
```

```python
#Average monthly income by education and attrition.
edu_attrition_income = df[["Education", "Attrition", "MonthlyIncome"]]
edu_attrition_income = edu_attrition_income.groupby(["Education", "Attrition"]).agg('mean')
print(edu_attrition_income)
edu_attrition_income.unstack(level=-1)['MonthlyIncome'].plot(ylabel='MonthlyIncome',kind='bar')
```
